# **Generative AI in three parts**
The composition of modern Generative AI can be broken down into three phases:
1.   **Pre-training** - Downloading the Internet and other large datasets (Analogy: grade school level background knowledge)
2.   **Post-training: Supervised Fine Tuning (SFT)** - A type of transfer learning that updates pre-training based on task specific data. (Analogy: college with a major)
3.   **Post-training: Reinforcement Learning** - Type of SFT. We will break this up into two types.
 - There are two types of reinforcement learning paradigms we will discuss:
   - **Human-based Reinforement Learning** - Uses humans (and now other Large Language Models (LLMs)) to vote on the correct prompt and answer pairs for GenAI chats to allow the models to learn from the human experience (Anology: Supervised on the job training).
    - **Full Reinforcement Learning** - Allows models to learn how to optimize their responses through experience based learning that is not human or LLM coded. The models are given an enviroment that has desired outcomes or goals but multiple paths or actionable options to reach those outcomes or goals (Anology: Non-Supervised on the job training).

##**Encoding Pretrained Data**
The output of the pretraining data filtering step is raw text. Now that we have (basically) downloaded the Internet and have raw text, we need to encode that data to be input into a neural network so statistical patterns can be discovered.

###**Encoding - Basics and Links to More Information**
Encoding involves representing data in a way that the neural networks can understand. Current LLM architectures rely on tokens for representation:

*   At the lowest level of computer encoding, symbols/tokens have a vocalulary of 1's and 0's (two possible sybmols)
*   Representing information into bytes (combinations of 8 1's and 0's, shrunk sequence length by 8 times and the vocabulary or symbol/token length would be 256 because that is the number of potential combinations of grouping 1's and 0's into groups of 8.
*   The byte-pair algorithm pairs consecutive pairs of bytes to create new possible tokens for encoding. This algorithm can be ran many times to create more tokens within the vocabulary and reducing the sequence length.
*   When you reduce sequence length and increase token vocabulary, you compress information more and lose some context.
*   For reference, OpenAI has a token vocabulary of 100K+ possible combinations.
*   The process of moving from raw bits to sequences of byte pairs (tokens) is called tokenization.

##**Pre-training Data Representation for Transfomer Input**
Now that we have encoded tokens that follow each other sequentially, we can input this into a neural network to be trained to identify the statistical patterns. Neural networks accept data in one dimensional sequences of symbols, this is a finite set so it needs to be efficient.

In the past (before 2017) the state of the art was called Recurrent Neural Networks (RNNs) or Long Short Term Memory (LSTM) Neural networks which captured token context from prevous states sequentially and statically. This type of context does not scale with large data.

In 2017 a seminal paper called "Attention is all you need" was published and introduced the Transformer network. The Transformer is a neural network which uses an appoach called "self attention". With this approach models make decisions on what should be generated by evaluating the sentence or phrase as a whole and not just one or two words in its history like previous appraoches. Additionally, attention networks have multiple "heads" that each focus on a separate aspect of the sentence or phrase, such as subject-verb agreement in one head and percentage of next word in another head.

*   You can find the paper here: [Attention is All You Need Paper](https://arxiv.org/pdf/1706.03762)

##**Train the Transfomer Neural Network**

###**Updating the Transformer to be an Optimal Model**
At this step, we need to model the statistical relationship of how the tokens follow each other in the sequence.

Training consists of matching the encoded tokens with parameters/weights. Since all the data is known and concatenated from the encoding in the pretraining step (byte/pair algorithm), the order is known and each following token acts as labled data.

*   The transformer architecture evaluates all tokens at one time to identify following tokens in parallel.

*   Initially the weights are random, but since the tokens are known and in seqeuential order, each following token acts as a label and the proceeding tokens that are sampled are evaluated many times to generalize and predict the lable.
*   Each evaluation is called an iterations of the data (epochs). The weights typically start out wrong so a loss function shows high loss.
*   After many epochs, the loss becomes minimal and now we have a generalized model that can perform inference.
*   Each epoch changes the parameters by improving the predictions of every token in the dataset at the same time.
*   The loss shows you how well you are doing in your training.
*   GPUs provide the wide scale parallization necessary to run modern LLM stacks.
*   Multiple GPUs collaborate to produce the next token.

##**Model Inference**
When you send prompts to ChatGPT, Claude, or any other LLM, you are performing inference.
Inference predicts one token at a time based on the token window.
- The key here is that the parameters (weights) are set at this point.
- The inference uses a separate neural network that samples from the distributions at training time. This can be thought of as a biased flipping of a coin.
- GPT-2 is the first time the full stack came together (Transformer neural network with 1.6 Billion parameters/weights, maximum context lenghth of 1024 tokens, trained on 100 Billion tokens.
- The context window shows the tokens that are used in making the next token prediction.
- Each epoch changes the parameters by improving the predictions of every token in the dataset at the same time.
- The loss shows you how well you are doing in your training.
- GPUs provide the wide scale parallization necessary to run modern LLM stacks.
- Multiple GPUs collaborate to produce the next token.

#**Post Training - Reinforcement Learning**

With reinforcement learning we teach the LLM to learn from experience. It will do trial and error until it comes up with the correct answer.
- The key is we don't have the final answer, like in supervised learning. We have the final desired result and the computer identifies the optimal way to achieve this result through trial and error.
- Many solutions are done in parallel to find the optimal solution efficiently.

Deepseek used reinformcent learning to get around the lack of access of GPUs and other hardware that provided massive parallization of algorithms.
- [Deepseek paper](https://arxiv.org/pdf/2501.12948)
- As models were updated through epochs, accuracy was improved
- Identified qualitative ways for the model to achieve better results
- Used more tokens to get higher accuracy (longer solutions) - thinking mode
- Focused on chain of thought (better to re-evaluate steps)
- This emerges in the optimization and is not hard coded anywhere.
